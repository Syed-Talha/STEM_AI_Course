#2.5: To SK we Go!

*sklearn is a library that operates on arrays, from which we use the preprocessing tool and the OneHotEncoder().*

````
import sklearn.preprocessing as skpre
import numpy as np
````

````
encoder = skpre.OneHotEncoder()  #creates a OneHotEncoder() object
````

*An array acts like a DataFrame without the labels (which are superfluous for ML), and we can access it using the values attribute.*

````
df_input.values
````

*We now feed our encoder information about the origin countries and number of cylinders, and we can check this info later. This is accomplished with the fit function.*

````
encoder.fit(X = df_input[['cylinders', 'origin']].values)  #must feed array, careful to double index
````

````
column_names = encoder.get_feature_names()
column_names    #checks for in order: EU, JPN, USA, 3 cyl, 4 cyl, 5 cyl, 6 cyl, 8 cyl
````

*We can now TRANSFORM this data into ones and zeroes via the transform function, which gives a matrix we convert into a multidimensional array.*

````
array = encoder.transform(X = df_input[['cylinders', 'origin']].values).toarray()
array
````

*DataFrames are more convenient, so lets turn the array into one. We get our row names from the input frame's index attribute, and the column names from our fitting during encoding.*

````
row_names = df_input.index
````

````
df_encoded = pd.DataFrame(data = array, index = row_names, columns = column_names)
df_encoded
````

*Now we just want to merge our new DataFrame with the original input frame!*

````
df_inputmerged = pd.concat(objs = [df_input, df_encoded], axis = 1)
df_inputmerged
````

````
df_inputfinal = df_inputmerged.drop(columns = ['origin', 'cylinders', 'name', 'is_usa','is_europe','is_japan'])
#remove all columns with string names now, and the redundant encoded ones
df_inputfinal
````
